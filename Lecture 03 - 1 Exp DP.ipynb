```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
values = numpy.random.standard_normal(10)
```

```python
bar = lambda x: plt.bar(range(len(x)), x)
```

```python
bar(values)
```

```python
max(values)
```

```python
q = - numpy.abs(values - max(values))
```

```python
bar(q)
```

```python
epsilon = 0.5
```

```python
pr = numpy.exp(epsilon*q) / numpy.sum(numpy.exp(epsilon*q))
bar(pr)
```

```python
epsilon = 0.2
pr = numpy.exp(epsilon*q) / numpy.sum(numpy.exp(epsilon*q))
bar(pr)
```

```python
epsilon = 2
pr = numpy.exp(epsilon*q) / numpy.sum(numpy.exp(epsilon*q))
bar(pr)
```

```python
X = ['red', 'yellow', 'green', 'blue']
p_X = numpy.random.uniform(0, 1, len(X))
p_X /= sum(p_X)
bar(p_X)
```

```python
data_cat = numpy.random.choice(X, size=100, p=p_X)
```

```python
data_cat
```

```python
from collections import Counter
```

```python
c = Counter(data_cat)
```

```python
c
```

```python
from statistics import mode
```

```python
most_common = mode(data_cat)
```

```python
most_common
```

```python
q = numpy.array([1 if o == most_common else 0 for o in X])
```

```python
q, X
```

```python
epsilon_cat = 0.5
pr_cat = numpy.exp(epsilon_cat*q) / sum(numpy.exp(epsilon_cat*q))
```

```python
pr_cat
```

```python
plt.bar(X, pr_cat)
```

```python
numpy.random.choice(X, p=pr_cat)
```
